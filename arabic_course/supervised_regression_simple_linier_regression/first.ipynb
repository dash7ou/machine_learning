```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([5,10,15,20,25]).reshape((-1,1))
y=np.array([12,20,28,38,42])
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
pre=model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,pre, 'r')
plt.show()
```
