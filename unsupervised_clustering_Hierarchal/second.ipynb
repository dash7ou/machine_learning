```python
# create model
```

```python
df = data.drop("country", axis=1)
```

```python
df
```

```python
Scal_df = StandardScaler().fit_transform(df)
Scal_df
```

```python
inertia = []
k = range(1,10)
for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit_transform(Scal_df)
    inertia.append(km.inertia_)
```

```python
plt.plot(k, inertia, "o-")
plt.title("Elbow Method For Optimal K")
```

```python
km = KMeans(n_clusters=7, random_state=42)
km.fit_transform(Scal_df)
km.inertia_
```

```python
labels=km.labels_
data['target']=labels
data.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as hcluster
```

```python
plt.figure(figsize=(50,15))
dendrogram = hcluster.dendrogram(hcluster.linkage(Scal_df, method='ward')) # ward, single, avarage
```

```python
hcl = AgglomerativeClustering(n_clusters=4, linkage='ward')
```

```python
hcl.fit_predict(Scal_df)
```

```python
hcl_label= hcl.labels_
data['target']=hcl_label
```
